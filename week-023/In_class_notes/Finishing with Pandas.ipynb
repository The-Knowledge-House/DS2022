# How can we quickly create derived columns?



```python
import pandas as pd
air_quality = pd.read_csv("https://raw.githubusercontent.com/pandas-dev/pandas/main/doc/data/air_quality_no2.csv", index_col=0,parse_dates=True)
```

```python
air_quality["london_mg_per_cubic"] = air_quality["station_london"] * 1.882
air_quality
```

```python
#Using this column initializer notation applies it element_wise. This is a sneaky way to get around having to use
#For loops.



air_quality["ratio_paris_antwerp"] = ( air_quality["station_paris"] / air_quality["station_antwerp"])
air_quality
```

```python
# Renaming columns


air_quality_renamed = air_quality.rename(columns={"station_antwerp":"BT3003"})
air_quality_renamed
```

```python
#Some Useful things to know
# Replace in the column name


air_quality_renamed.columns = air_quality_renamed.columns.str.replace("_","!")
air_quality_renamed
```

# How do we calculate some stats?




```python
titanic = pd.read_csv("https://raw.githubusercontent.com/pandas-dev/pandas/main/doc/data/titanic.csv")
```

```python
titanic["Age"].mean()
```

```python
titanic[["Age","Fare"]].median()
```

```python
titanic.describe()
```

```python
titanic.agg({"col_name":["list of values"]})
#https://pandas.pydata.org/docs/user_guide/basics.html#basics-stats
```

```python
# We know about filtering, but how about grouping. Pandas is of the split-apply-combine pattern.
```

```python
titanic[["Sex","Age"]]
```

```python
titanic[["Sex","Age"]].groupby("Sex")
```

```python
titanic[["Sex","Age"]].groupby("Sex").mean()
```

```python
titanic.groupby("Sex").mean()
```

```python
titanic.groupby("Sex")["Age"].mean()
```

```python
titanic.groupby(["Sex","Pclass"])["Fare"].mean()
```

```python
titanic["Pclass"].value_counts()
```

```python
titanic.groupby(["Pclass","Sex"])["Pclass"].count()
```

# Making plots in Pandas

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


ts = pd.Series(np.random.randn(1000), index=pd.date_range("1/1/2000", periods=1000))
ts
```

```python
ts = ts.cumsum()
ts
```

```python
height = [3, 12, 5, 18, 45]
bars = ('Men', 'Women')
x_pos = np.arange(len(bars))

plt.bar(x_pos, height, color = (0.5, 0.1, 0.7, 0.6))

plt.title('Titanic demos')
plt.xlabel('Women and men')
plt.ylabel('Survival rate')

plt.xticks(x_pos, bars)

plt.show()
```
